# 機械学習（クラスター分析）Demo

## Iris dataset (アヤメの計測データ、通称：アイリス)
“setosa”, “versicolor”, “virginica” という 3 種類の品種のアヤメのがく片 (Sepal)、花弁 (Petal) の幅および長さを計測したデータです。

- sepal length (cm)	がく片の長さ
- sepal width (cm)	がく片の幅
- petal length (cm)	花弁の長さ
- petal width (cm)	花弁の幅

![](https://archive.ics.uci.edu/ml/assets/MLimages/Large53.jpg)
出典： https://archive.ics.uci.edu/ml/datasets/Iris

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
%matplotlib inline

iris = datasets.load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

## ペアプロット図（散布図行列）

```python
sns.pairplot(df)
plt.grid(True)
```

## クラスター分析（４次元データ、３つのクラスターに分類）

```python
from sklearn import cluster, datasets

k_means = cluster.KMeans(n_clusters=3)
k_means.fit(iris.data) 
```

```python
# クラスター分析で３分類した結果で色分け
plt.title("Clustering")
plt.scatter(iris.data[:, 0], iris.data[:, 1], c=k_means.labels_) 
plt.xlabel("sepal length (cm)")
plt.ylabel("sepal width (cm)")
plt.grid(True)
```

```python
# クラスター分析で３分類した結果で色分け
plt.title("Clustering")
plt.scatter(iris.data[:, 2], iris.data[:, 1], c=k_means.labels_) 
plt.xlabel("petal length (cm)")
plt.ylabel("sepal width (cm)")
plt.grid(True)
```

```python
# クラスター分析で３分類した結果で色分け
plt.title("Clustering")
plt.scatter(iris.data[:, 3], iris.data[:, 1], c=k_means.labels_) 
plt.xlabel("petal width (cm)")
plt.ylabel("sepal width (cm)")
plt.grid(True)
```

```python
# 元データの品種で色分け
plt.title("original species.")
plt.scatter(iris.data[:, 0], iris.data[:, 1], c=iris.target) 
plt.xlabel("sepal length (cm)")
plt.ylabel("sepal width (cm)")
plt.grid(True)
```

```python
# 元データの品種で色分け
plt.title("original species.")
plt.scatter(iris.data[:, 2], iris.data[:, 1], c=iris.target) 
plt.xlabel("petal length (cm)")
plt.ylabel("sepal width (cm)")
plt.grid(True)
```

```python
# 元データの品種で色分け
plt.title("original species.")
plt.scatter(iris.data[:, 3], iris.data[:, 1], c=iris.target) 
plt.xlabel("petal width (cm)")
plt.ylabel("sepal width (cm)")
plt.grid(True)
```

## 主成分分析で２次元化

```python
from sklearn import decomposition
pca = decomposition.PCA(n_components=2)
pca.fit(iris.data)
```

```python
X = pca.transform(iris.data)
```

```python
# クラスター分析で３分類した結果で色分け
plt.scatter(X[:, 0], X[:, 1], c=k_means.labels_) 
plt.grid(True)
```

```python
# 元データの品種で色分け
plt.scatter(X[:, 0], X[:, 1], c=iris.target) 
plt.grid(True)
```
