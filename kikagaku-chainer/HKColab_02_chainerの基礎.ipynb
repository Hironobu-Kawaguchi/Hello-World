### 実践問題

#### 問題設定
入力`[1,2,3]`に対する出力`y`の値を計算して下さい。

#### 構成
- 左側のリンク: fc1 →(3,2)
- 右側のリンク: fc2 →(2,1)
- fc1の線形変換 →u1
- fc1の非線形変換(Relu) →z1
- fc2の線形変換 →出力y

```
np.random.seed(3)
x = np.array([[1,2,3]],'f')
fc1 = L.Linear(3,2)
fc2 = L.Linear(2,1)
u1 = fc1(x)
z1 = F.relu(u1)
y = fc2(z1)
y
```

```
# 解答
np.random.seed(3)

fc1 = L.Linear(3,2)
fc2 = L.Linear(2,1)

x = np.array([[1,2,3]],'f')

u1 = fc1(x)
z1 = F.relu(u1)

y = fc2(z1)
y
```

## 損失関数の計算

```
#教師データ
t = np.array([[3]],'f')
t
```

```
#平均二乗誤差の計算
loss = F.mean_squared_error(t,y)
loss
```
